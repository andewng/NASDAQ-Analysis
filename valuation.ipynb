```python
import yfinance as yf
```

In this notebook, the focus will be on stock analysis, valuation using different ML/Visualization Techniques

```python
def stock_info(ticker):
    stock = yf.Ticker(ticker)
    hist = stock.history(period="5y")
    cashflow = stock.cashflow
    balance_sheet = stock.balance_sheet
    info = stock.info
    return hist, cashflow, balance_sheet, info
```

```python
stock_info("MSTR")
```

```python
def fetch_data(ticker):
    stock = yf.Ticker(ticker)
    hist = stock.history(period="5y")
    return hist
```

```python
MSTR = fetch_data("MSTR")
MSTR.head()
```

```python
MSTR.describe()
```

```python
import plotly.graph_objects as go
```

```python
# Closing Price
fig = go.Figure(data=[go.Candlestick(
    x=MSTR.index,
    open=MSTR['Open'],
    high=MSTR['High'],
    low=MSTR['Low'],
    close=MSTR['Close'],
    name='MSTR'
)])

fig.update_layout(
    title='Closing Price of MSTR over Time',
    xaxis_title='Date',
    yaxis_title='Price (USD)',
    template='plotly_white',
    width=900,
    height=400,
    xaxis_rangeslider_visible=False  # removes the range slider for a cleaner look
)

fig.show()
```

```python
# Trading Volume

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=MSTR.index,
    y=MSTR['Volume'],
    mode='lines',
    name='Trade Volume',
    line=dict(color='blue')
))

fig.update_layout(
    title='Trade Volume of MSTR over Time',
    xaxis_title='Date',
    yaxis_title='Volume',
    template='plotly_white',
    width=800,
    height=400,
    xaxis_rangeslider_visible=False
)

fig.show()
```

```python
MSTR['20_MA'] = MSTR['Close'].rolling(window=20).mean()
MSTR['50_MA'] = MSTR['Close'].rolling(window=50).mean()
MSTR['200_MA'] = MSTR['Close'].rolling(window=200).mean()
```

```python
# moving average
fig = go.Figure()

# Closing Price
fig.add_trace(go.Scatter(
    x=MSTR.index,
    y=MSTR['Close'],
    mode='lines',
    name='Closing Price',
    line=dict(color='blue')
))

# 20-Day Moving Average
fig.add_trace(go.Scatter(
    x=MSTR.index,
    y=MSTR['20_MA'],
    mode='lines',
    name='20-Day MA',
    line=dict(color='red')
))

# 50-Day Moving Average
fig.add_trace(go.Scatter(
    x=MSTR.index,
    y=MSTR['50_MA'],
    mode='lines',
    name='50-Day MA',
    line=dict(color='orange')
))

# 200-Day Moving Average
fig.add_trace(go.Scatter(
    x=MSTR.index,
    y=MSTR['200_MA'],
    mode='lines',
    name='200-Day MA',
    line=dict(color='green')
))

# Layout
fig.update_layout(
    title='MSTR Stock Price with Moving Averages',
    xaxis_title='Date',
    yaxis_title='Price (USD)',
    template='plotly_white',
    width=800,
    height=500
)

fig.show()
```
